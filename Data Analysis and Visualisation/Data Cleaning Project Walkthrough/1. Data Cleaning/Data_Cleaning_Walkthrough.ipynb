### Inserting DBN Fields
<font size='3'>

The `class_size` doesn't have a `DBN` column and `hs_directory` does have a `dbn` column, which requires renaming.

From looking at the both datasets rows, the `DBN` in the `sat_results` data is just a combination of the `CSD` and `SCHOOL CODE` columns in the `class_size` data.
The main difference is that the `DBN` is padded, so that the `CSD` portion of it always consists of two digits. We'll add a leading `0` to the `CSD` if the `CSD` is less than two digits long. 

```python
#Copy the dbn column to DBN
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]
```

```python
#function to create padded number for 1 digit number(below 10)
def pad_csd(num):
    #Fill the string with zeros until it is 2 characters long
    return str(num).zfill(2)
```

```python
pad_csd(3)
```

```python
#apply the function to change the 'CSD' column to padded CSD
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad_csd)
```

```python
#combine two cilumns to recreate the 'DBN' column for 'class_size' dataset
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]
print(data["class_size"].head())
```

### Combining the SAT Scores
<font size='3'>

Create a new column that totals up the SAT scores for the different sections of the exam. 
For that :
- convert the `SAT Math Avg. Score`, `SAT Critical Reading Avg. Score`, and `SAT Writing Avg. Score` columns in the `sat_results` data set to a numeric data type.

```python
#convert from string to numeric the `errors="coerce"` argument treats any invalid strings it can't convert 
#to numbers as missing values instead
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']
for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce")
```

```python
# add up sat score values from three columns 
data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]
print(data['sat_results']['sat_score'].head())
```

### Parsing Geographic Coordinates for Schools
<font size='3'>

Next, we'll want to parse the latitude and longitude coordinates for each school. This will enable us to map the schools and uncover any geographic patterns in the data. 

```python
data["hs_directory"]["Location 1"].head()
```

```python
#function with regular expression to extract lattitude
import re

def find_lat(loc):
    coords = re.findall("\(.+\)", loc)
    lat = coords[0].split(",")[0].replace("(", "")
    return lat
```

```python
#extract latitude from `Location 1` column
data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(find_lat)

print(data["hs_directory"].head())
```

### Extracting the Longitude

```python
#function with regular expression to extract longitude
import re

def find_lon(loc):
    coords = re.findall("\(.+\)", loc)
    lon = coords[0].split(",")[1].replace(")", "").strip()
    return lon
```

```python
#extract longitude from `Location 1` column
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(find_lon)
data["hs_directory"].head()
```

```python
data["hs_directory"].info()
```

```python
#convert to numeric values
data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors="coerce")

print(data["hs_directory"].head())
```
